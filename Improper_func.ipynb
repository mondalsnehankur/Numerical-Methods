```python
import matplotlib.pyplot as plt
import numpy as np
import random
from scipy.integrate import simps
```

```python
def g(t):
  return np.exp(-t**2)

def MonteCarlo(a,x,N,g):
  t = np.random.uniform(a,x,N)
  return (x-a)*np.mean(g(t))
```

```python
N = 100
X = [0.25,0.5,1.0,2.0]
```

```python
for x in X:
  print("X = ",np.round(x,4),"erf(x)=",np.round(MonteCarlo(0,x,N,g),4))
```
