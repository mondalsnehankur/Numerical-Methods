# ***Monte Carlo Integration.***

*In mathematics, Monte Carlo integration is a technique for numerical integration using random numbers. It is a particular Monte Carlo method that numerically computes a definite integral. While other algorithms usually evaluate the integrand at a regular grid, Monte Carlo randomly chooses points at which the integrand is evaluated. This method is particularly useful for higher-dimensional integrals.*

MonteCarloIntegrationCircle.svg.png

````python
import random
import numpy as np
import matplotlib.pyplot as plt
````



```
# Initialisation.
```



````python
X = []
Y = []
A = []
B = []
````



```
# Function
```



````python
def pi_val(n):
  m = 0
  for i in range(n):
    x = random.random()
    y = random.random()
    z = x**2 + y**2
    if z <= 1.0:
      X.append(x)
      Y.append(y)
      m += 1
    else:
      A.append(x)
      B.append(y)
  return(4.0*m/n,X,Y,A,B)
````



```
# Plotting
```



````python
plt.scatter(pi_val(10000)[1],pi_val(10000)[2],label='+',color='grey')
plt.scatter(pi_val(10000)[3],pi_val(10000)[4],label='*',color='black')

plt.text(0.2,0.22," • Area under the Curve")

plt.title("Moñte Carlo Integration.")
plt.axhline(0,linestyle='--',color='black')
plt.axvline(0,linestyle='--',color='black')
plt.legend(loc='upper right')
plt.grid()

plt.show()
````
